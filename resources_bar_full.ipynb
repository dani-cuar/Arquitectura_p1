```python
import matplotlib.pyplot as plt
label = ["Tamaño L1I","Tamaño L1D","Tamaño L2","Tamaño L3","Entradas ROB","Entradas BTB","Numero de ALUs","Ancho Pipeline"]

for i in range(len(data)):
    nombres_tamanos = data[i][0]
    valores_cpi_mp3_enc = data[i][1]
    valores_cpi_h264_dec = data[i][2]

    # Crear dos subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Configurar el primer subplot (valores_cpi_mp3_enc)
    bars1 = ax1.bar(nombres_tamanos, valores_cpi_mp3_enc, color='skyblue')
    ax1.set_title('Workload mp3_enc')
    ax1.set_xlabel(label[i])
    ax1.set_ylabel('Valor CPI')
    ax1.set_ylim(1.20,1.21)

    # Configurar el segundo subplot (valores_cpi_h264_dec)
    bars2 = ax2.bar(nombres_tamanos, valores_cpi_h264_dec, color='lightcoral')
    ax2.set_title('Workload h264_dec')
    ax2.set_xlabel(label[i])
    ax2.set_ylabel('Valor CPI')
    ax2.set_ylim(0.88,0.97)

    # Añadir los valores de las barras al primer subplot
    for bar in bars1:
        height = bar.get_height()
        ax1.annotate(f'{height:.4f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                    textcoords='offset points', ha='center', va='bottom')

    # Añadir los valores de las barras al segundo subplot
    for bar in bars2:
        height = bar.get_height()
        ax2.annotate(f'{height:.4f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                    textcoords='offset points', ha='center', va='bottom')

    # Ajustar el espaciado entre subplots
    plt.tight_layout()
    
    plt.savefig('grafico_'+label[i]+'.png')
    
    # Mostrar el gráfico
    plt.show()
```
